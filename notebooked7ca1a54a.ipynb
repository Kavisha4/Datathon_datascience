```python
grades =[]
for i in data['Percentage']:
   if i>90 and i<100:
      grades.append('S')
   elif i>80 and i<=90:
      grades.append('A')
   elif i>70 and i<=80:
      grades.append('B')
   elif i>60 and i<=70:
      grades.append('C')
   elif i>40 and i<=60:
      grades.append('D')
   else:
      grades.append('F')
data['Grades'] = grades
data
```
